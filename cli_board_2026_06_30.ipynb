```python
"""
## 개선 필요 사항
  Q. title과 content 리스트가 2개임
     → 하나로 묶어서 더 효율적으로 수정
     → board list([title, content])
     → 딕셔너리 사용?
  S1. reshape 사용해서 차원 변경

======== CLI 게시판 실행 ========
명령어 ) article write (사용자 입력)
제목 : 123 (사용자 입력)
내용 : abc (사용자 입력)
1번 글이 생성 되었습니다.
명령어 ) article write (사용자 입력)
제목 : 456 (사용자 입력)
내용 : def (사용자 입력)
2번 글이 생성 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
 1      /   123     /   abc
================================

명령어 ) exit (사용자 입력)
======== CLI 게시판 종료 ========

# 핵심함수: input()
# 추가된 사항: article list
   - 데이터 저장: 리스트.append()
   - 딕셔너리가 효과적까?
# 예상되는 예외사항
   -
"""
board_list = []  # 게시판 리스트 = [제목, 내용]
board_count = 0  # 생성된 게시판 수를 저장

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    # strip(): 공백 제거
    command = input("명령어 ) ").strip()

    # 입력된 명령어 일치 여부 확인
    if command == "article write":
        # 제목, 내용을 리스트에 순차적으로 추가 후 reshape
        board_list.append(input("제목: ").strip())
        board_list.append(input("내용: ").strip())
        # 게시글이 생성될 때마다 1씩 증가
        board_count += 1
        print("{}번 글이 생성되었습니다.".format(board_count))

    elif command == "article list":
        print("===========================")
        print("번호 /   제목   /   내용   ")
        # board_list의 크기를 다차원 배열 (board_count, 2)로 변환
        result = np.reshape(board_list, (board_count, 2))
        # 생성된 게시글의 역순으로 출력
        for i in range(board_count, 0, -1):
            print("{}    /   {}    /   {}   ".format(i,
                        result[board_count-1, 0], result[board_count-1, 1]))
        print("===========================")

    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break

    else:
        print("올바른 명령어(article write, article list, exit)를 입력하세요")
```
