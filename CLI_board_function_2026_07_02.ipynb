```python
"""
TODO/DONE:
    article dic → article class로 변경

"""
from datetime import datetime
from dataclasses import dataclass

# Article 데이터 클래스 정의
@dataclass
class Article:
    """게시글 정보를 저장하는 데이터 클래스

    Attributes:
        id (int): 게시글 번호
        reg_date (str): 등록 일자
        update_date (str): 변경 일자
        head (str): 게시글 제목
        body (str): 게시글 내용
    """

    id: int
    reg_date: str
    update_date: str
    head: str
    body: str


# 회원정보 클래스 정의
class Member(Article):
    """회원 정보를 관리하는 클래스: Article 클래스 상속

    Attributes:
        login_id (str):
        login_pw (str):
        name (str) : 회원명
    Method:
        member
    """

    def __init__(self, login_id, login_pw, name):
        self.login_id = login_id
        self.login_pw = login_pw
        self.name = name

    def member_join(self):
        """회원가입"""
        pass

    def login_id_check(self):
        """아이디 중복검사"""
        pass

    def login_pw_check(self):
        """비밀번호 검사"""
        pass


# delete 함수 정의
def article_delete(boards, id, number):
    """ 게시글을 삭제(delete)
    Args:
        command (string) : 입력된 명령어
        boards (list) : 게시글(atricle) 저장,
                        atricle은 딕셔너리 {'id': 번호,
                        'head': 제목,
                        'body': 내용,
                        'time': 생성 시간}
        id (list) : 전체 게시글의 번호를 저장
    """
    # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
    try:
        # index 기반 삭제
        del boards[id.index(number)]  # boards.remove(article) → object 기반 삭제
        print("{}번 글이 삭제 되었습니다.".format(number))
    except:
        print("{}번 글은 존재하지 않습니다.".format(number))

# edit 함수 정의
def article_edit(boards, id, number):
    """ 게시글을 수정(edit)
    Args:
        상동
    """
    # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
    try:
        a = id.index(number)
        title = input("제목: ").strip()
        content = input("내용: ").strip()
        # 게시글의 제목과 내용 수정
        boards[a].update(head = title, body = content)
        print("{}번 글이 수정 되었습니다.".format(number))
    except:
        print("{}번 글은 존재하지 않습니다.".format(number))

# detail 함수 정의
def article_detail(boards, id, number):
    """ 게시글의 상세내용(detail) 출력
    Args:
        상동
    """
    # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
    try:
        # 게시글 선택
        article = boards[id.index(number)]
        # 게시글 상세 출력
        print("번호 : {}".format(number))
        print("날짜 : {}".format(article.get('time')))
        print("제목 : {}".format(article.get('head')))
        print("내용 : {}".format(article.get('body')))
    except:
        print("{}번 글은 존재하지 않습니다.".format(number))

# 게시글의 기능들을 선택해서 실행하는 함수 정의
def board_modification(command, boards, id):
    """ 게시글의 기능들을 선택해서 실행
    Args:
        command (string) : 입력된 명령어
        boards (list) : 게시글(atricle) 저장,
                        atricle은 딕셔너리 {'id': 번호, 'head': 제목, 'body': 내용}
        id (list) : 전체 게시글의 번호를 저장
    """

    command_number = command.split()
    # 번호가 아닐 경우 ValueError 발생
    try:
        cmd = command_number[1]
        number = int(command_number[2])

        if not boards:
            print("게시글이 없습니다.")
        else:
            # 현재 게시되어 있는 글들의 번호 저장
            # 매번 전체글을 호출해서 번호를 추출
            # 현재는 풀스캔 방법이 맞음, 향후 DBMS 적용
            for article in boards:
                value = list(article.values())
                id.append(value[0])

            # 함수 호출
            match cmd:
                case 'delete':
                    article_delete(boards, id, number)
                case 'edit':
                    article_edit(boards, id, number)
                case 'detail':
                    article_detail(boards, id, number)
                case _:  # startswith에서 조건에 맞는 명령어만 입력
                    pass
    except:
        print("숫자를 입력하세요")

# article 테스트 데이터 생성
def make_test_data(boards, number=3):
    """ 게시글 자동 생성
    Args:
        number (int) : 자동으로 생성할 수
        boards (list) : 게시글(atricle) 저장,
                        atricle은 딕셔너리 {'id': 번호, 'head': 제목, 'body': 내용}
    """

    reg_time = datetime.now().strftime('%Y-%m-%d')

    for count in range(number):
    # 게시글 내용 입력
        article = Article(count+1, reg_time, reg_time, 'auto_head', 'auto_content')
        # article = {'id': i+1 , 'head': 'auto_title_{}'.format(i+1),
        #            'body': 'auto_content_{}'.format(i+1), 'time': reg_time}
        boards.insert(0, article)
    return count


"""===================================================================="""


created_article_count = make_test_data(boards)  # 생성된 게시글 수

boards = []  # 딕셔너리('id':번호, 'head':제목, 'body': 내용)을 저장하는 리스트
id = []  # 게시글 번호 저장

# 프로그램 시작 시 게시글 자동 생성: defalut 3개
make_test_data(boards, 3)

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    # # append는 값을 계속 추가하기 때문에 초기화 필요
    id.clear()
    # split()로 입력된 데이터를 분리
    command = input("명령어 ) ").strip()

    # 입력된 명령어 일치 여부 확인
    if command == "article write":
        title = input("제목: ").strip()
        content = input("내용: ").strip()
        reg_time = datetime.now().strftime('%Y-%m-%d')
        # 게시글이 생성될 때마다 1씩 증가
        created_article_count += 1
        # 게시글 내용 입력
        article = Article(created_article_count, reg_time, reg_time, title, content)
        # article = {'id': created_article_count, 'head': title, 'body': content, 'time': reg_time}
        # append는 순차적으로 증가, insert로 첫 번째 저장
        boards.insert(0, article)
        print("{}번 글이 생성되었습니다.".format(created_article_count))

    elif command == "article list":
        if not boards:
            print("게시글이 없습니다.")
        else:
            print("=====================================")
            print(" 번호 /   제목   /   내용   /   날짜 ")

            for article in boards:
                # 딕셔너리의 value 값을 리스트에 저장
                # result = list(article.values())
                print("  {}   /   {}    /   {}  /  {}".format(article.id,
                                    article.head, article.body, article.reg_date))
            print("=====================================")

    elif command.startswith('article delete'):
        # 게시판 수정 함수 호출: delete 기능 수행
        board_modification(command, boards, id)

    elif command.startswith('article edit'):
        # 게시판 수정 함수 호출: edit 기능 수행
        board_modification(command, boards, id)

    elif command.startswith('article detail'):
        # 게시판 수정 함수 호출: detail 기능 수행
        board_modification(command, boards, id)

    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break
    else:
        print("지원하지 않는 기능입니다")
```
