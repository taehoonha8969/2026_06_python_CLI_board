```python
from dataclasses import dataclass
from datetime import datetime

@dataclass
class Article:
    """게시글 정보를 저장"""
    num: int
    title: str
    body: str
    id: str
    date: str

@dataclass
class Member:
    """회원 정보를 저장"""
    num: int
    id: str
    pw: str
    name: str
    date: str

def date():
    now = datetime.now()
    current_date = now.strftime('%Y.%m.%d %H:%M')
    return current_date

# 테스트용 게시글 생성
def initial_article_for_test(articles):
    num = 3

    for i in range(num, 0, -1):
        article = Article(i, 'title{}'.format(i), 'test{}'.format(i**2), 'test{}'.format(i), date())
        # article = Article(num=i, title='test', body='test', id='admin', date=date())
        articles.append(article)

    # return이 없으면 값이 반환이 안됨
    return articles

# 테스트용 회원 정보 생성
def initial_member_for_test(members):
    num = 3

    for i in range(num, 0, -1):
        member = Member(i, 'test{}'.format(i), 'aaa', 'hooney', date())
        members.append(member)

    return members

# 게시글 및 회원 정보 저장
articles = []
members = []

# 최종 등록된 게시글 수와 회원 수
last_article_count = 3
last_member_count = 3
# 로그인된 아이디 저장
logined_id = None

# 상태 저장 변수
check_id = True
check_pw = True
check_login = True
login_status = False

initial_article_for_test(articles)
initial_member_for_test(members)

print('========== CLI Board Start ==========')

while True:
    command = input("Command: ").strip()

    if command == 'exit':
        break

    # 회원 등록: ID 중복여부 확인 및 PW 2번 입력 동일여부 확인
    elif command == 'member join':
        # ID 중복 여부 확인
        while check_id:
            id_join = input('ID: ').strip()
            for member in members:
                if id_join == member.id:
                    print('\"{}\" is already existed. Please input another ID'.format(id_join))
                    break

            # 입력된 ID가 회원 정보에 없을 때 break
            if id_join != member.id:
                check_id == False
                logined_id = member.id
                break

        # PW 일치 여부 확인
        while check_pw == True:
            pw_join_1 = input('PW1: ').strip()
            pw_join_2 = input('PW2: ').strip()

            # 입력된 두 PW가 서로 일치하지 않으면 다시 입력
            if pw_join_1 != pw_join_2:
                print('Password is not matched. Please input password again.')
                check_pw = True
            else:
                check_pw = False
                break

        last_member_count += 1  # ID가 없고 PW가 일치할 경우 회원 수 증가
        login_status = True  # 회원 가입이 완료되었을 때 로그인 상태를 True로
        name = input('name: ').strip()
        member = Member(last_member_count, id_join, pw_join_1, name, date())
        members.insert(0, member)  # 회원 정보를 제일 처음에 저장

    # 회원 로그인: ID 및 PW 일치 여부 확인
    elif command == 'member login':
        while check_login:
            id_login = input('ID: ').strip()
            pw_login = input('PW: ').strip()

            for member in members:
                # ID가 존재하고 PW가 맞을 경우 break
                if id_login == member.id:
                    if pw_login == member.pw:
                        print('login success!!!')
                        logined_id = member.id
                        check_login = False
                        break
            # ID or PW is not matched.
            if (id_login != member.id) or (pw_login != member.pw):
                print('ID or PW is not matched, please input again.')

        # ID가 존재하고 PW가 맞을 경우 login 상태를 True로 설정
        login_status = True

    # 회원 로그아웃 (로그인된 상태에서 로그아웃 가능)
    elif command == 'member logout':
        if login_status == True:
            print('Logout success!!!')
            login_status = False
        else:
            print('You are not logined, please login first.')

    # 회원 정보 출력(option for test)
    elif command == 'member list':
        print(' # / ID / PW / Name / Date ')
        for m in members:
            print('{} / {} / {} / {} / {}'.format(m.num, m.id, m.pw, m.name, m.date))

    # 게시글 입력 (로그인된 상태에서 작성 가능)
    elif command == 'article write':
        last_article_count += 1

        if login_status == True:
            title = input('Title: ').strip()
            body = input('Body: ').strip()
            article = Article(last_article_count, title, body, logined_id, date())
            articles.insert(0, article)

    # 게시글 출력 (로그인된 상태에서 확인 가능)
    elif command == 'article list':
        if login_status == True:
            print(' # / Title / Body / ID / Date ')
            for a in articles:
                print('{} / {} / {} / {} / {}'.format(a.num, a.title, a.body, a.id, a.date))
        else:
            print('You are not logined, please login first.')

    # 게시글 삭제 (로그인된 상태에서 삭제 가능)
    elif command.startswith('article delete'):
        print('Delete Test')



    # 게시글 수정 (로그인된 상태에서 수정 가능)
    elif command.startswith('article edit'):
        pass

    else:
        print("Please input correct command.")

print('=========== CLI Board End ===========')
```
