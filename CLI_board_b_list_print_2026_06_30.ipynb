```python
"""
======== CLI 게시판 실행 ========
명령어 ) article write (사용자 입력)
제목 : 123 (사용자 입력)
내용 : abc (사용자 입력)
1번 글이 생성 되었습니다.
명령어 ) article write (사용자 입력)
제목 : 456 (사용자 입력)
내용 : def (사용자 입력)
2번 글이 생성 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
 1      /   123     /   abc
================================

명령어 ) exit (사용자 입력)
======== CLI 게시판 종료 ========

# 핵심함수: input()
# 추가된 사항: article list
   - 데이터 저장: 리스트.append()
# 예상되는 예외사항
   -

"""
title = []  # 제목을 저장할 리스트
content = []  # 내용을 저장할 리스트
board_count = 0  # 생성된 게시판 수를 저장

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    command = input("명령어 ) ")
    # 명령어 일치여부 확인
    if command == "article write":
        title.append(input("제목: "))
        content.append(input("내용: "))
        # 게시글이 생성될 때마다 1씩 증가
        board_count += 1
        print("{}번 글이 생성되었습니다.".format(board_count))
    elif command == "article list":
        print("===========================")
        print("번호 /   제목   /   내용   ")
        # 생성된 게시글의 역순으로 출력
        for i in range(board_count, 0, -1):
            print("{}    /   {}    /   {}   ".format(i, title[i - 1], content[i - 1]))
        print("===========================")
    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break
    else:
        print("올바른 명령어(article write, article list, exit)를 입력하세요")
```
