```python
"""
4. 게시글 수정 기능 구현: article edit #
  - 딕셔너리 value 수정: update 함수 ex) dic.update('key': 'value')
5. 게시글 상세 보기 구현: article detail #


======== CLI 게시판 실행 ========
명령어 ) article write (사용자 입력)
제목 : 123 (사용자 입력)
내용 : abc (사용자 입력)
1번 글이 생성 되었습니다.
명령어 ) article write (사용자 입력)
제목 : 456 (사용자 입력)
내용 : def (사용자 입력)
2번 글이 생성 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
 1      /   123     /   abc
================================
명령어 ) article delete 3 (사용자 입력)
3번 글은 존재하지 않습니다.
명령어 ) article delete 1 (사용자 입력)
1번 글이 삭제 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
================================
명령어 ) exit (사용자 입력)
======== CLI 게시판 종료 ========

"""
created_article_count = 0  # 생성된 게시글 수
deleted_article_count = 0  # 삭제된 게시글 수, 현재 사용 X

boards = []  # 게시글을 저장하는 딕셔너리 (번호, 제목, 내용)

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    # split()로 입력된 데이터를 분리
    command = input("명령어 ) ").strip()

    # 입력된 명령어 일치 여부 확인
    if command == "article write":
        # 제목, 내용을 리스트에 순차적으로 추가 후 reshape
        title = input("제목: ").strip()
        content = input("내용: ").strip()
        # 게시글이 생성될 때마다 1씩 증가
        created_article_count += 1
        article = {'id': created_article_count, 'title': title, 'content': content}
        # append는 순차적으로 증가, insert는 원하는 위치에 삽입
        # 첫 번째로 저장
        boards.insert(0, article)
        print("{}번 글이 생성되었습니다.".format(created_article_count))

    elif command == "article list":
        print("===========================")
        print(" 번호 /   제목   /   내용   ")

        for article in boards:
            # 딕셔너리의 value 값을 리스트에 저장
            result = list(article.values())
            print("  {}   /   {}    /   {}".format(result[0], result[1], result[2]))
        print("===========================")

    elif command.startswith('article delete'):
        check_number = command.split()
        # 번호가 아닐 경우 문자열로 인해서 에러 발생
        try:
            delete_num = int(check_number[2])
            if delete_num > created_article_count:
                print("{}번 글은 존재하지 않습니다.".format(delete_num))

            elif (0 < delete_num) and (delete_num <= created_article_count):
                del_articles = boards
                del_count = 0

                for article in del_articles:
                    result = list(article.values())
                    # value가 delete_num와 같으면 그 행을 삭제
                    if result[0] == delete_num:
                        del boards[del_count]
                        deleted_article_count += 1
                        print("{}번 글이 삭제 되었습니다.".format(delete_num))
                    del_count += 1
            elif delete_num == 0:
                print("게시글이 없습니다.")
        except:
            print("숫자를 입력하세요")
    elif command.startswith('article edit'):
        check_number = command.split()
        try:
            edit_num = int(check_number[2])
            if edit_num > created_article_count:
                print("{}번 글은 존재하지 않습니다.".format(delete_num))

            elif (0 < edit_num) and (edit_num <= created_article_count):

                del_count = 0

                for article in del_articles:
                    result = list(article.values())

    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break
    else:
        print("지원하지 않는 기능입니다")
```
