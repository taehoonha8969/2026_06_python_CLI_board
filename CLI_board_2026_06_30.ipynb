```python
"""
## 해결책: 딕셔너리 사용 → 딕셔너리를 리스트로 저장

## 개선 필요 사항
  Q. title과 content 리스트가 2개임
     → 하나로 묶어서 더 효율적으로 수정
     → board list([title, content])
     → 딕셔너리 사용?
  S2. 딕셔러니 사

======== CLI 게시판 실행 ========
명령어 ) article write (사용자 입력)
제목 : 123 (사용자 입력)
내용 : abc (사용자 입력)
1번 글이 생성 되었습니다.
명령어 ) article write (사용자 입력)
제목 : 456 (사용자 입력)
내용 : def (사용자 입력)
2번 글이 생성 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
 1      /   123     /   abc
================================

명령어 ) exit (사용자 입력)
======== CLI 게시판 종료 ========

# 핵심함수: input()
# 추가된 사항: article list
   - 데이터 저장: 리스트.append()
   - 딕셔너리가 효과적까?
# 예상되는 예외사항
   -
"""
articles = []  # 게시판 딕셔너리 (게시글 수, 제목, 내용)
board_count = 0  # 생성된 게시판 수를 저장

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    # strip(): 공백 제거
    command = input("명령어 ) ").strip()

    # 입력된 명령어 일치 여부 확인
    if command == "article write":
        # 제목, 내용을 리스트에 순차적으로 추가 후 reshape
        title = input("제목: ").strip()
        content = input("내용: ").strip()
        # 게시글이 생성될 때마다 1씩 증가
        board_count += 1
        board = {'id': board_count, 'title': title, 'content': content}
        articles.append(board)
        print("{}번 글이 생성되었습니다.".format(board_count))

    elif command == "article list":
        print("===========================")
        print(" 번호 /   제목   /   내용   ")
        # 생성된 게시글의 역순으로 출력exit
        # len(boards) = 3 일텐데???
        for article in articles:
            # 딕셔너리의 value 값을 리스트에 저장
            result = list(article.values())
            print("  {}   /   {}   /   {}".format(result[0], result[1], result[2]))
        print("===========================")

    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break

    else:
        print("올바른 명령어(article write, article list, exit)를 입력하세요")
```
