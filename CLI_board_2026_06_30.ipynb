```python
"""
5. 게시글 상세 보기 구현: article detail #
번호 :
날짜 :
제목 :
내용 :

======== CLI 게시판 실행 ========
명령어 ) article write (사용자 입력)
제목 : 123 (사용자 입력)
내용 : abc (사용자 입력)
1번 글이 생성 되었습니다.
명령어 ) article write (사용자 입력)
제목 : 456 (사용자 입력)
내용 : def (사용자 입력)
2번 글이 생성 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
 1      /   123     /   abc
================================
명령어 ) article delete 3 (사용자 입력)
3번 글은 존재하지 않습니다.
명령어 ) article delete 1 (사용자 입력)
1번 글이 삭제 되었습니다.
명령어 ) article list
================================
번호    /   제목    /   내용
 2      /   456     /   def
================================
명령어 ) exit (사용자 입력)
======== CLI 게시판 종료 ========

"""
from datetime import datetime

created_article_count = 0  # 생성된 게시글 수

boards = []  # 딕셔너리('id':번호, 'head':제목, 'body': 내용)을 저장하는 리스트
id = []  # 게시글 번호 저장

print("======== CLI 게시판 실행 ========")

# exit 명령어를 입력 받을 때까지 실행
while True:
    # # append는 값을 계속 추가하기 때문에 초기화 필요
    id.clear()
    # split()로 입력된 데이터를 분리
    command = input("명령어 ) ").strip()

    # 입력된 명령어 일치 여부 확인
    if command == "article write":
        title = input("제목: ").strip()
        content = input("내용: ").strip()
        # 게시글이 생성될 때마다 1씩 증가
        created_article_count += 1
        # 게시글 내용 입력
        article = {'id': created_article_count, 'head': title, 'body': content}
        # append는 순차적으로 증가, insert로 첫 번째 저장
        boards.insert(0, article)
        print("{}번 글이 생성되었습니다.".format(created_article_count))

    elif command == "article list":
        if not boards:
            print("게시글이 없습니다.")
        else:
            print("===========================")
            print(" 번호 /   제목   /   내용   ")

            for article in boards:
                # 딕셔너리의 value 값을 리스트에 저장
                result = list(article.values())
                print("  {}   /   {}    /   {}".format(result[0], result[1], result[2]))
            print("===========================")

    elif command.startswith('article delete'):
        check_number = command.split()
        # 번호가 아닐 경우 문자열로 인해서 에러 발생
        try:
            delete_num = int(check_number[2])

            if not boards:
                print("게시글이 없습니다.")
            else:
                # 현재 게시되어 있는 글들의 번호 저장
                for article in boards:
                    value = list(article.values())
                    id.append(value[0])

                # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
                try:
                    del boards[id.index(delete_num)]
                    print("{}번 글이 삭제 되었습니다.".format(delete_num))
                except:
                    print("{}번 글은 존재하지 않습니다.".format(delete_num))
        except:
            print("숫자를 입력하세요")

    elif command.startswith('article edit'):
        check_number = command.split()
        # 번호가 아닐 경우 문자열로 인해서 에러 발생
        try:
            edit_num = int(check_number[2])

            if not boards:
                print("게시글이 없습니다.")
            else:
                # 현재 게시되어 있는 글들의 번호 저장
                for article in boards:
                    value = list(article.values())
                    id.append(value[0])
                # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
                try:
                    a = id.index(edit_num)
                    title = input("제목: ").strip()
                    content = input("내용: ").strip()
                    # 게시글의 제목과 내용 수정
                    boards[a].update(head = title, body = content)
                    print("{}번 글이 수정 되었습니다.".format(edit_num))
                except:
                    print("{}번 글은 존재하지 않습니다.".format(edit_num))
        except:
            print("숫자를 입력하세요")

    elif command.startswith('article detail'):
        check_number = command.split()
        # 번호가 아닐 경우 문자열로 인해서 에러 발생
        try:
            detail_num = int(check_number[2])

            if not boards:
                print("게시글이 없습니다.")
            else:
                # 현재 게시되어 있는 글들의 번호 저장
                for article in boards:
                    value = list(article.values())
                    id.append(value[0])
                # index()의 찾는 값이 없을 경우 ValueError를 발생하기 때문에 예외처리 필요
                try:
                    a = id.index(detail_num)
                    head = boards[a].get('head')
                    body = boards[a].get('body')
                    # 게시글 상세 출력
                    print("번호 : {}".format(detail_num))
                    print("날짜 : {}".format(datetime.now()))
                    print("제목 : {}".format(head))
                    print("내용 : {}".format(body))
                except:
                    print("{}번 글은 존재하지 않습니다.".format(detail_num))
        except:
            print("숫자를 입력하세요")

    elif command == "exit":
        print("======== CLI 게시판 종료 ========")
        break
    else:
        print("지원하지 않는 기능입니다")
```
